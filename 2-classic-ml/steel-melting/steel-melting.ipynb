# Проект "Промышленность"

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Вам предстоит построить модель, которая предскажет температуру стали.

**Описание этапа обработки**

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

**Описание данных**

Данные состоят из файлов, полученных из разных источников:

- `data_arc.csv` — данные об электродах;
- `data_bulk.csv` — данные о подаче сыпучих материалов (объём);
- `data_bulk_time.csv` *—* данные о подаче сыпучих материалов (время);
- `data_gas.csv` — данные о продувке сплава газом;
- `data_temp.csv` — результаты измерения температуры;
- `data_wire.csv` — данные о проволочных материалах (объём);
- `data_wire_time.csv` — данные о проволочных материалах (время).

Во всех файлах столбец `key` содержит номер партии. В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

Проведем графический анализ группами добавок со схожим количеством применений

```python
def histogram(data, column, num_list):
    for num in num_list:
        data[column+' '+str(num)].plot(kind='hist', label=column+' '+str(num), alpha=0.6, bins=100, figsize=(8,6))

    plt.legend()
    plt.title('Распределение '+column)
    plt.xlabel('Объём')
    plt.ylabel('Количество значений')
    plt.show()
```

### Работа с датасетом data_arc

По плану нужно посчитать суммарное время нагрева партии, средние реактивную и активную мощности по партии, их сумму, а также посчитать затраты энергии на партию путём умножения времени на суммарную мощность. Поехали.

```python
## приводим временные данные к соответствующему типу
for column in ['Начало нагрева дугой', 'Конец нагрева дугой']:
    data_arc[column] = pd.to_datetime(data_arc[column], format='%Y-%m-%d %H:%M:%S')
    
data_arc.info()
```

Удачно. Считаем время нагрева в секундах

```python
data_arc['heat_time'] = data_arc['Конец нагрева дугой'] - data_arc['Начало нагрева дугой']
data_arc['heat_time'] = data_arc['heat_time'].dt.total_seconds()
data_arc.head()
```

Превосходно. Теперь временные отметки нам не нужны, вдальнейшем от них избавимся (и не нужно переименовывать).
Считаем суммарную мощность во время нагрева. После группировки она станет средней, поэтому выбрано имя avg_power

```python
data_arc['avg_power'] = data_arc['Активная мощность'] + data_arc['Реактивная мощность']

data_arc.head()
```

Далее группируем данные по партии, применяя запланированные функции.

```python
agg_func = {'avg_power':'mean',
            'heat_time':'sum',
            'Активная мощность':'mean',
            'Реактивная мощность':'mean'}

data_arc_agg = data_arc.groupby(['key']).agg(agg_func)
data_arc_agg.head()
```

Чтож, прошло удачно. Теперь посчитаем затраченную энергию.

```python
data_arc_agg['spent_energy'] = data_arc_agg['avg_power'] * data_arc_agg['heat_time']
data_arc_agg.head()
```

Группировка успешно завершена. Переходим к целевому признаку

## Подбор гиперпараметров моделей

### LinearRegression

Страшная по величине ошибка. Как себя покажут другие модели? Но для начала напишем функцию для подбора гиперпараметров.

### Функция для подбора гиперпараметров

```python
def best_model(model, hyperparams):
    # готовим и обучаем классификатор
    clf = GridSearchCV(model, hyperparams, scoring='neg_mean_absolute_error', n_jobs=-1, cv=5, verbose=1)
    clf.fit(features_train, target_train)
    
    # получаем результаты подбора и записываем в переменные время обучения и предсказания
    results = pd.DataFrame(clf.cv_results_)
    fit_time = results[results['rank_test_score'] == 1]['mean_fit_time'].values[0]
    predict_time = results[results['rank_test_score'] == 1]['mean_score_time'].values[0]
    
    # записываем в переменные лучшую модель, её параметры и RMSE
    model = clf.best_estimator_
    params = clf.best_params_
    mae = clf.best_score_   
    
    # выводим результат
    print(f'Лучшая модель {model}')
    print(params)
    print()
    print(f'Показатель MAE: {mae}')
    print()
    print(f'Время обучения: {fit_time}; Время предсказания: {predict_time}')
    return model, params, mae, fit_time, predict_time
```

### DecisionTreeRegressor

```python
model_dtr = DecisionTreeRegressor(random_state=RANDOM_STATE)
hyperparams = {'max_depth':[x for x in range(5, 20)]}

model_dtr, dtr_params, dtr_mae, dtr_fit_time, dtr_predict_time = best_model(model_dtr, hyperparams)
```
