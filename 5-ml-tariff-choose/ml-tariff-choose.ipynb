# Рекомендация тарифов

В нашем распоряжении данные о поведении клиентов, которые уже перешли на тарифы "Смарт" и "Ультра" оператора мобильной связи "Мегалайн". На их основе нужно построить модель для задачи классификации, которая выберет подходящий конкретному пользователю тариф. Известно, что предобработка данных не понадобится — её уже сделали.

Доля правильных ответов построенной модели должна быть не менее 75%

## Описание данных

Каждый объект в наборе данных — это информация о поведении одного пользователя за месяц. Известно:

- `сalls` — количество звонков,
- `minutes` — суммарная длительность звонков в минутах,
- `messages` — количество sms-сообщений,
- `mb_used` — израсходованный интернет-трафик в Мб,
- `is_ultra` — каким тарифом пользовался в течение месяца («Ультра» — 1, «Смарт» — 0).

## Импорт библиотек

Прежде, чем приступить к изучению данных, импортируем необходимые библиотеки

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## Откройте и изучите файл

Так как проект выполняется локально, используем конструкцию `try - except`, чтобы при загрузке файла на ревью, не возникло проблем с путём к файлу, а также применим метод `info`, чтобы убедиться в отсуствии пропусков, корректности типов данных и визуально взглянем на 10 случайных строк таблицы:

```python
try:
    users = pd.read_csv('/datasets/users_behavior.csv')
except:
    users = pd.read_csv('users_behavior.csv')
    
display(users.sample(10))
```

Визуально всё выглядит корректно

```python
users.info()
```

Пропусков не обнаружено, типы вполне себе подходящие, названия столбцов в порядке.

## Разбейте данные на выборки

Нам необходимо три выборки. Обучающая, валидационная и тестовая. Объёмы выборок в процентном соотношении должны быть соответственно 60%, 20% и 20%. Сделаем это последовательно:

```python
# создаём датасет со свойствами
user_features = users.drop(['is_ultra'], axis=1)
# создаём датасет с целевыми признаками
tariff_target = users['is_ultra']

# создаём обучающий датасет объёмом 60% от исходного
features_train, features_valid, target_train, target_valid = train_test_split(
    user_features, tariff_target, test_size=0.4, random_state=12345)
# оставшуюся часть делим пополам и получаем части по 20% от исходного объёма
features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid, target_valid, test_size=0.5, random_state=12345)

# проверим, как резделены данные по объёму
for data in [features_train, features_valid, features_test, target_train, target_valid, target_test]:
    print(f'Размерность: {data.shape}')    
```

Видно, что целевые соотношения объёмов выборок получены - 1928 объектов в обучающей, и по 643 в валидационной и тестовой выборках. Проверим визуально первые две строки выборок:

```python
for data in [features_train, features_valid, features_test]:
    display(data.head(2))
```

Сравнивая первые строки получившихся выборок делаем вывод, что они разные.

## Исследуйте модели

Исследование моделей будем проводить с помощью цикла, перебирающего значения гиперпараметров, выводя показатели точности каждой итерации.

### Дерево решений

Для дерева решений перебираем гиперпараметр "максимальная глубина":

```python
# создаем переменную для модели с лучшим показателем точности
best_tree_model = None
# создаем переменную для лучшего показателя точности
best_tree_accuracy = 0
# создаем переменную для значения максимальной глубины, обеспечившей максимальную точность
best_tree_depth = 0

# пробежимся по значениям глубины от 1 до 10
for depth in range(1, 11):
    model_tree = DecisionTreeClassifier(max_depth=depth, random_state=12345) # создаём дерево
    model_tree.fit(features_train, target_train) # тренируем на тренировочной выборке
    prediction_tree_valid = model_tree.predict(features_valid) # предсказываем результаты по валидационным свойствам
    accuracy_tree_valid = accuracy_score(target_valid, prediction_tree_valid) # считаем точность
    print(f'Точность дерева решений при максимальной глубине {depth}: {accuracy_tree_valid}')
    # условной конструкцией записываем в ранее созданные переменные лучшие показатели
    if accuracy_tree_valid > best_tree_accuracy:
        best_tree_model = model_tree
        best_tree_accuracy = accuracy_tree_valid
        best_tree_depth = depth

print() # отделим лучший результат от промежуточных
print('Лучшая модель дерева решений:')
print(f'- Глубина: {best_tree_depth}')
print(f'- Точность: {(best_tree_accuracy * 100).round(2)}%')
```

Видим, что минимальное значение точности оказалось при глубине 1 - 75%. При остальных значениях - 77-78%. Лучшая модель с глубиной 3 показала точность 78,54%

### Случайный лес

Для случайного леса аналогичным способом перебираем максимальную глубину и количество деревьев в модели.

```python
# создаём переменные для хранения лучших показателей
best_forest_model = None
best_forest_accuracy = 0
best_forest_est = 0
best_forest_depth = 0

# циклом перебираем количество деревьев от 10 до 50 включительно с шагом 10, глубину так же - от 1 до 10
for est in range(10, 51, 10):
    for depth in range(1, 11):
        model_forest = RandomForestClassifier(max_depth=depth, n_estimators=est, random_state=12345)
        model_forest.fit(features_train, target_train)
        accuracy_forest_valid = model_forest.score(features_valid, target_valid)
        print(f'Точность случайного леса из {est} деревьев при максимальной глубине {depth}: {accuracy_forest_valid}')
        if accuracy_forest_valid > best_forest_accuracy:
            best_forest_model = model_forest
            best_forest_accuracy = accuracy_forest_valid
            best_forest_depth = depth
            best_forest_est = est

print() # отделим лучший результат от промежуточных
print('Лучшая модель случайного леса:')
print(f'- Количество деревьев: {best_forest_est}')
print(f'- Глубина: {best_forest_depth}')
print(f'- Точность: {(best_forest_accuracy * 100).round(2)}%')
```

Видно, что количество деревьев в выбранном диапазоне не особо влияет на точность исследования - максимальная точность варьируется находится в диапазоне 80-81% - прирост по сравнению со случайным деревом в 1,5 - 2%. Скорее сам факт использования модели случайного леса даёт увеличение точности.

### Логистическая регрессия

Для логистической регрессии аналогично перебираем количество итераций:

```python
best_logistic_model = None
best_logistic_accuracy = 0
best_logistic_iter = 0

# перебираем от 100 до 1000 с шагом 100
for ite in range(100, 1000, 100):
    model_logistic = LogisticRegression(max_iter = ite, solver='lbfgs', random_state=12345)
    model_logistic.fit(features_train, target_train)
    accuracy_logistic_valid = model_logistic.score(features_valid, target_valid)
    print(f'Точность логистической регрессии при {ite} итерациях: {accuracy_logistic_valid}')
    
    if accuracy_logistic_valid > best_logistic_accuracy:
        best_logistic_model = model_logistic
        best_logistic_accuracy = accuracy_logistic_valid
        best_logistic_iter = ite

print() # отделим лучший результат от промежуточных
print('Лучшая модель логистической регрессии:')
print(f'- Количество итераций: {best_logistic_iter}')
print(f'- Точность: {(best_logistic_accuracy * 100).round(2)}%')
```

Никакой зависимости точности от количества итераций не видно... Предполагаю, что это максимально возможная точность данной модели.

### Итоги исследования

- Максимальную точность на валидационной выборке показала модель случайного леса - 80,87%
- Минимальная точность у логистической регрессии - 71,07%
- Дерево решений по точности ближе к случайному лесу, поэтому думаю, что для бинарной классификации модель логистической регрессии не подходит.

## Проверьте модель на тестовой выборке

Проверим модели на тестовой выборке, заодно сравним точности предсказаний по всем выборкам (и обучающей тоже):

### Тест лучшего дерева решений

```python
# посчитаем результат лучшей модели дерева решений на обучающей выборке
train_tree_prediction = best_tree_model.predict(features_train)
train_tree_accuracy = accuracy_score(target_train, train_tree_prediction)

# применяем модель и считаем точность на тестовой выборке
best_tree_prediction = best_tree_model.predict(features_test)
test_tree_accuracy = accuracy_score(target_test, best_tree_prediction)

print(f'Точность предсказаний решающего дерева по тестовой выборке: {(test_tree_accuracy * 100).round(2)}%')
print(f'Точность предсказаний решающего дерева по валидационой выборке: {(best_tree_accuracy * 100).round(2)}%')
print(f'Точность предсказаний решающего дерева по обучающей выборке: {(train_tree_accuracy * 100).round(2)}%')
```

Судя по результатам на тестовой и обучающей выборках, модель очень даже обучена.

### Тест лучшего случайного леса

```python
# посчитаем результат лучшей модели случайного леса на обучающей выборке
train_forest_accuracy = best_forest_model.score(features_train, target_train)

# применяем модель и считаем точность на тестовой выборке
test_forest_accuracy = best_forest_model.score(features_test, target_test)

print(f'Точность предсказаний случайного леса по тестовой выборке: {(test_forest_accuracy * 100).round(2)}%')
print(f'Точность предсказаний случайного леса по валидационой выборке: {(best_forest_accuracy * 100).round(2)}%')
print(f'Точность предсказаний решающего дерева по обучающей выборке: {(train_forest_accuracy * 100).round(2)}%')
```

Разрыв в точности между обучающей и тестовой выборкой приличный, но результат на тесте показан достаточно высокий. Модель слегка переобучена.

### Тест логистической регрессии

```python
# посчитаем результат лучшей модели на обучающей выборке
train_logistic_accuracy = best_logistic_model.score(features_train, target_train)
# применяем модель и считаем точность на тестовой выборке
test_logistic_accuracy = best_logistic_model.score(features_test, target_test)

print(f'Точность предсказаний логистической регрессии по тестовой выборке: {(test_logistic_accuracy * 100).round(2)}%')
print(f'Точность предсказаний логистической регрессии по валидационой выборке: {(best_logistic_accuracy * 100).round(2)}%')
print(f'Точность предсказаний решающего дерева по обучающей выборке: {(train_logistic_accuracy * 100).round(2)}%')
```

Результат был низкий, а стал ещё хуже. Низкие результаты и по тестам и в обучении - модель недообучена (или необучаема в данных условиях?)

### Итоги теста лучших моделей

- Лучшая модель - случайный лес: 79,63% точности. При этом модель переобучена
- Худшая модель - логистическая регрессия: 68,43% точности. Модель недообучена.

    - Дерево решений - на полпроцента
    - Случайный лес - на процент

## (бонус) Проверьте модели на адекватность

Судя по теоретическим материалам, проверить модель на адекватность нужно сравнить результаты её применения со случайным выбором. Так как мы имеем дело с бинарной классификацией, нужно понять, какова вероятность случайно попасть в тариф "Ультра". И простым вычитанием из единицы мы получим вероятность попасть в тариф "Смарт".

```python
# считаем вероятность случайно выбрать тариф "Ультра"
ultra_prob = target_test.sum() / target_test.count()
# считаем вероятность случайно выбрать тариф "Смарт"
smart_prob = 1 - ultra_prob
print(f'Вероятность попасть в "Ультра" в тестовой выборке: {(ultra_prob * 100).round(2)}%')
print(f'Вероятность попасть в "Смарт" в тестовой выборке: {(smart_prob * 100).round(2)}%')
```

Таким образом мы понимаем, что если мы при определении тарифов будем **всегда** выбирать "Смарт", то будем правы в 68,43% случаев. Это значение и будет границей адекватности. Точность нашей модели должна быть ощутимо выше, чтобы было целесообразно её применять. Проверим наши модели на адекватность и сделаем выводы в автоматическом режиме (программисты мы или где? :-)):

```python
# создадим словарь с ключами-моделями и значениями-точностями в тесте
models_dict = {
    'Логистическая регрессия' : test_logistic_accuracy ,
    'Случайный лес'  : test_forest_accuracy,
    'Дерево решений' : test_tree_accuracy}

# пробежимся циклом по словарю, сравнивая точность модели в тесте со случайным выбором.
for model, accuracy in models_dict.items():
    if smart_prob == accuracy:
        print(f'{model} жульничает!')
    elif smart_prob > accuracy:
        print(f'{model} вредит!')
    else:
        print(f'{model} круче случайного выбора на {((accuracy - smart_prob) * 100).round(2)}%')
```

Точность предсказаний логистической регрессии в точности равна вероятности попасть в тариф "Смарт" - **68,43%**. Вывод: **регрессия - жулик**! Проставила везде "Смарт". Модель явно неадекватна.

А вот дерево решений и случайный лес показали себя лучше, и мы вполне можем их применять.
