#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. 

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Построим модель для такого предсказания.

Для этого нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

Радует, что пропусков нет. Однако, необходимо сменить тип данных в столбце `datetime` для работы с временными рядами. Взглянем визуально, как выглядит запись даты для использования метода `to_datetime`

## Подготовка

Сменим тип данных и проконтролируем результат.

```python
taxi['datetime'] = pd.to_datetime(taxi['datetime'], format='%Y-%m-%d %H:%M:%S')

taxi.info()
```

Это успех :) Теперь присвоим индексам значения из столбца `datetime`, отсортируем их по возрастанию и сразу произведем ресемплирование по одному часу.

```python
taxi = taxi.set_index('datetime')
taxi.sort_index(inplace=True)
taxi = taxi.resample('1H').sum()

print(taxi.info())
taxi.head()
```

Судя по выведенным данным запланированные действия выполнены: индексы в виде даты, тип нужный, и по первые пять записей явно по возрастанию, а значит и остальные.

Напишем функцию для создания признаков: 
- день недели
- час
- отстающие значения
- скользящее среднее

Другие характерные для временных рядов признаки (год, месяц, число) не имеет смысла использовать, учитывая горизонт прогнозирования в один час и выборку, включающую только 6 месяцев

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

Отлично, осталось определить количество столбцов с "отстающими значениями" и количество часов для расчёта скользящего среднего. Думаю, что разумно выбрать для смещения два часа для отслеживания краткосрочного тренда и учёта количества заказов в предыдущий час, а скользящее среднее взять 24 часа, чтобы оценивать среднее количество заказов за сутки.

Создадим признаки и проконтролируем результат визуально.

```python
make_features(taxi, 2, 24)

taxi.head(30)
```

Отлично. С пропущенными значениями разберемся при подготовке к обучению

## Подбор модели и гиперпараметров

### Разделение датасета на выборки

Подготовим выборки для обучения: разделим подготовленный датасет в соотношении 90% на 10%, а также удалим строки с пропусками из тренировочной выборки. Проконтролируем индексы.

```python
train, test = train_test_split(taxi, shuffle=False, test_size=0.1)
train = train.dropna()

print(train.index.min(), train.index.max())
print(test.index.min(), test.index.max())
```

Всё в порядке - тренировочная выборка заканчивается 13 августа в 13-00, тестовая начинается в тот же день в 14-00.

Выделим свойства и целевой признак с контролем размерности результата

```python
features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']

features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']

for data in [features_train, features_test, target_train, target_test]:
    print(data.shape)
```

Разделение произошло корректно.

### Функция для подбора параметров и оценки модели

Во избежание дублирования кода, напишем функцию, которая принимает на вход модель и гиперпараметры для перебора, а на выходе даёт лучшую модель, её гиперпараметры, RMSE лучшей модели, а также время обучения и предсказания (интересно же). Обратим внимание на специфику временного ряда - тестовая выборка должна быть позже предыдущей. Поэтому воспользуемся кроссвалидатором для временных рядов TimeSeriesSplit. Количество сплитов зададим 5, т.к. в тренировочной выборке осталось пять целых месяцев, поэтому каждая тестовая (внутри CV) выборка будет чуть больше месяца.

```python
def best_model(model, hyperparams):
    # задаём кросс-валидатор
    tscv = TimeSeriesSplit(n_splits=5)
    # готовим и обучаем классификатор
    clf = GridSearchCV(model, hyperparams, scoring='neg_root_mean_squared_error', n_jobs=-1, cv=tscv, verbose=1)
    clf.fit(features_train, target_train)
    
    # получаем результаты подбора и записываем в переменные время обучения и предсказания
    results = pd.DataFrame(clf.cv_results_)
    fit_time = results[results['rank_test_score'] == 1]['mean_fit_time'].values[0]
    predict_time = results[results['rank_test_score'] == 1]['mean_score_time'].values[0]
    
    # записываем в переменные лучшую модель, её параметры и RMSE
    model = clf.best_estimator_
    params = clf.best_params_
    rmse = clf.best_score_   
    
    # выводим результат
    print(f'Лучшая модель {model}')
    print(params)
    print()
    print(f'Показатель RMSE: {rmse}')
    print()
    print(f'Время обучения: {fit_time}; Время предсказания: {predict_time}')
    return model, params, rmse, fit_time, predict_time
```

Готово! Теперь рассмотрим модели:
- LinearRegression
- DecisionTreeRegressor
- CatBoostRegressor
- LGBMRegressor

### LinearRegression

Линейную регрессию оценим без подбора параметров, использовав функцию `cross_val_score`. Оценивать эту модель будем только по RMSE, т.к. скорее всего она вдребезги проиграет моделям градиентного бустинга. Те модели оценим по времени работы.

```python
model_lr = LinearRegression()
tscv = TimeSeriesSplit(n_splits=5)
lr_crossval_rmse = cross_val_score(model_lr, features_train, target_train, cv=tscv, scoring='neg_root_mean_squared_error')
lr_rmse = round(lr_crossval_rmse.mean(), 2)
print(f'Средний показатель RMSE линейной регрессии (кросс-валидация): {lr_rmse}')
```

### DecisionTreeRegressor

```python
model_dtr = DecisionTreeRegressor(random_state=12345)
hyperparams = {'max_depth':[x for x in range(5, 20)]}

model_dtr, dtr_params, dtr_rmse, dtr_fit_time, dtr_predict_time = best_model(model_dtr, hyperparams)
```

### CatBoostRegressor

```python
model_cbr = CatBoostRegressor(random_state=12345, verbose=False) 
hyperparams = [{'learning_rate':[0.1, 0.5, 0.8]}]

model_cbr, cbr_params, cbr_rmse, cbr_fit_time, cbr_predict_time = best_model(model_cbr, hyperparams)
```

### LGBMRegressor

```python
model_lgbmr = LGBMRegressor(random_state=12345)
hyperparams = [{'num_leaves':[31, 100, 200], 
                'learning_rate':[0.1, 0.3, 0.5]}]

model_lgbmr, lgbmr_params, lgbmr_rmse, lgbmr_fit_time, lgbmr_predict_time = best_model(model_lgbmr, hyperparams)
```
