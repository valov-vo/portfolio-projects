# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. 
    
**Наша задача — установить параметры.** Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных:**
* airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
* balcony — число балконов
* ceiling_height — высота потолков (м)
* cityCenters_nearest — расстояние до центра города (м)
* days_exposition — сколько дней было размещено объявление (от публикации до снятия)
* first_day_exposition — дата публикации
* floor — этаж
* floors_total — всего этажей в доме
* is_apartment — апартаменты (булев тип)
* kitchen_area — площадь кухни в квадратных метрах (м²)
* last_price — цена на момент снятия с публикации
* living_area — жилая площадь в квадратных метрах (м²)
* locality_name — название населённого пункта
* open_plan — свободная планировка (булев тип)
* parks_around3000 — число парков в радиусе 3 км
* parks_nearest — расстояние до ближайшего парка (м)
* ponds_around3000 — число водоёмов в радиусе 3 км
* ponds_nearest — расстояние до ближайшего водоёма (м)
* rooms — число комнат
* studio — квартира-студия (булев тип)
* total_area — площадь квартиры в квадратных метрах (м²)
* total_images — число фотографий квартиры в объявлении

## Предобработка данных

### Работа со столбцом locality_name

Здесь стоит пояснить, почему предобработку начинаем именно со столбца locality_name. Если сразу выявим неявные дубликаты здесь, то очистка от явных дубликатов пройдёт эффективнее.

```python
estate_data['locality_name'].unique()
```

Зоркий глаз аналитика заметил только один неявный дубликат 'поселок Пансионат Зелёный Бор' и 'посёлок Пансионат Зелёный Бор'. Произведем замену и проверим её успешность

```python
estate_data['locality_name'] = estate_data['locality_name'].replace('поселок Пансионат Зелёный Бор', 'посёлок Пансионат Зелёный Бор')
estate_data[estate_data['locality_name'] == 'поселок Пансионат Зелёный Бор']['locality_name'].count()
```

Порядок. От неявных дубликатов в столбце locality_name избавились.

Перейдём к удалению строк с пропусками и проверим результат

```python
estate_data = estate_data.dropna(subset=['locality_name']).reset_index(drop=True)
estate_data.info()
```

Количество ненулевых записей в столбце равно количеству записей в датафрейме. Значит, пропуски удалены, да и индексы в порядке

### Заполнение пропусков в столбце ceiling_height и kitchen_area

Изначально была мысль, заполнить пропуски в столбце kitchen_area <b>медианным значением</b>. Однако, смутило наличие столбца is_apartment. Подумал это что-то вроде студии и тогда отдельной кухни в таких квартирах не было бы и пропуски заполнили бы <b>нулём</b>. Оказалось, что апартаменты это не студия, а скорее юридический статус - жилое помещение по документам для проживания не предназначенное. 
<br>Поэтому сделал вывод, что апартаменты с точки зрения площади кухни выделяться не будут. Так что можно смело заполнять пропуски в столбцах ceiling_height и kitchen_area <b>медианным значением</b>.

```python
# считаем медианные значения
ceiling_height_median = estate_data['ceiling_height'].median()
kitchen_area_median = estate_data['kitchen_area'].median()
# заполняем пропуски медианными значениями
estate_data['ceiling_height'] = estate_data['ceiling_height'].fillna(ceiling_height_median)
estate_data['kitchen_area'] = estate_data['kitchen_area'].fillna(kitchen_area_median)
# выведем медианнные значения на экран для контроля
print(f'Медианная высота потолков: {ceiling_height_median}')
print(f'Медианная площадь кухни: {kitchen_area_median}')
estate_data.info()
```

Пропусков в столбце ceiling_height не обнаружено. Кстати, метод инфо подтвердил, что first_day_exposition содержит данные нужного нам формата

### Работа со столбцом living_area

В этом столбце наверняка есть корреляция между жилой площадью и количеством комнат. Поэтому заполнять пропуски в нем стоит с учётом количества комнат. Для начала посмотрим а сколькикомнатные квартиры присутствуют в датафрейме

```python
estate_data['rooms'].value_counts()
```

Бросается в глаза значение 0. Попробуем взглянуть поближе на эти строки

```python
estate_data[estate_data['rooms'] == 0].sort_values(by='living_area').head()
```

Судя по значениям жилой площади, можно включить эту группу данных в пул однокомнатных квартир, заменив 0 на 1

```python
estate_data.loc[estate_data['rooms'] == 0, 'rooms'] = 1
print(estate_data[estate_data['rooms'] == 0]['rooms'].count()) ## проверяем, есть ли ещё нулевые комнаты
```

Замена произведена. Двигаемся дальше. Чтобы адекватно произвести замену пропусков в соответствии с количеством комнат, думаю, стоит посчитать медианы отдельно для 1, 2, 3, 4 комнатных квартир, и группы квартир "5 комнат и выше". Иначе говоря, провести категоризацию, укрупнив группы.

```python
## напишем функцию для расчёта медиан в зависимости от числа комнат
def living_area_median(rooms):
    median = estate_data.loc[estate_data['rooms'] == rooms, 'living_area'].median()
    return median
## Создадим список для хранения медиан и циклом заполним его значениями в зависимости от числа комнат
medians = []
for rooms in range(1, 5):
    median = living_area_median(rooms)
    medians.append(median)
    
## Отдельно посчитаем медиану для категории "5 комнат и выше"
median_more_5 = estate_data.loc[estate_data['rooms'] >= 5, 'living_area'].median()
print(f'Список медиан для 1,2,3,4-хкомнатных квартир: {medians}')
print(f'Медиана для категории "5 комнат и выше": {median_more_5}')
```

Данные для замены пропусков подготовлены. Осуществим её и проверим результат.

```python
estate_data.loc[estate_data['rooms'] >= 5, 'living_area'] = estate_data.loc[estate_data['rooms'] >= 5, 'living_area'].fillna(median_more_5)
estate_data.loc[estate_data['rooms'] >= 5]['living_area'].value_counts().head() 
```

Видно, что произведено 49 замен. Теперь разберемся с остальными значениями rooms.

```python
for rooms in range(1, 5):
    estate_data.loc[estate_data['rooms'] == rooms, 'living_area'] = estate_data.loc[estate_data['rooms'] == rooms, 'living_area'].fillna(medians[rooms-1])
estate_data[estate_data.isna() == True]['living_area'].count() 
```

Пропусков нет. Для контроля корректности замены взглянем на значения жилой площади однокомнатных квартир.

```python
estate_data.loc[estate_data['rooms'] == 1]['living_area'].value_counts().head()
```

```python
# Переводим метры в километры, округляем и переводим в тип int
estate_data_clean['cityCenters_nearest_km'] = estate_data_clean['cityCenters_nearest'] / 1000
estate_data_clean['cityCenters_nearest_km'] = estate_data_clean['cityCenters_nearest_km'].round().astype('int')
# Делаем выборку по названию населённого пункта, отбрасывая нулевые значения расстояния до центра
st_petersburg_data = estate_data_clean.query('(locality_name == "Санкт-Петербург") and (cityCenters_nearest_km > 0)')
# Создаём таблицу со средней ценой квартиры с группировкой по расстоянию в километрах
st_petersburg_data_km = st_petersburg_data.pivot_table(index='cityCenters_nearest_km', values='last_price', aggfunc='mean').reset_index()
# Считаем стоимость километра и добавляем её в новый столбец
st_petersburg_data_km['km_cost'] = st_petersburg_data_km['last_price'] / st_petersburg_data_km['cityCenters_nearest_km']
st_petersburg_data_km.plot(
    title='Цена километра удалённости от центра',
    x='cityCenters_nearest_km',
    y='km_cost',
    grid=True,
    style='o-'
)
```

График вышел достаточно плавный. Возьмём границу центра в 5 км. Хотя если делать подобные действия на нефильтрованной базе. Граница видна на расстоянии в 3 км.
